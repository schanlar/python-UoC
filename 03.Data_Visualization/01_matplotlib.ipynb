- **Πολλαπλές καμπύλες σε ένα γράφημα**

```python
# Create the data for fitting curves
y1_fit_values = np.sin(x_values)
y2_fit_values = np.cos(x_values)
```

```python
plt.figure(figsize=(8,3))
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Voltage [V]', fontsize=16)
plt.title('Voltage vs Time', fontsize=16)

# Plot only the data points
_ = plt.plot(x_values, y1_values, marker='o', ms=5, ls=' ', c='red', label='sin data')
_ = plt.plot(x_values, y2_values, marker='o', ms=5, ls=' ', c='blue', label='cos data')

# Plot the fit curves
_ = plt.plot(x_values, y1_fit_values, ls='--', lw=2, c='red', label='sin (fit)')
_ = plt.plot(x_values, y2_fit_values, ls='-.', lw=2, c='blue', label='cos (fit)')

_ = plt.legend(loc='lower left', fontsize=15)
```

### Παράδειγμα

```python
# This is to make random data reproducable
np.random.seed(0)

# Generate some data
res_a1 = 0.2*np.random.randn(1000)+0.4
res_b1 = 0.25*np.random.randn(1000)+0.4
res_a2 = 0.21*np.random.randn(1000)+0.3
res_b2 = 0.22*np.random.randn(1000)+0.3
```

```python
# Good habit to come up with complicated strings beforehand
textstr1 = '\n'.join((
    r'$\sigma_a=%.4f$' % (np.std(res_a1)),
    r'$\sigma_b=%.4f$' % (np.std(res_b1))))

textstr2 = '\n'.join((
    r'$\sigma_a=%.4f$' % (np.std(res_a2)),
    r'$\sigma_b=%.4f$' % (np.std(res_b2))))
```

```python
# Create two plots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,3.5))


# First plot
ax1.hist(res_a1, bins=30, density=True, histtype='step', label='Method a', color='blue')
ax1.hist(res_b1, bins=30, density=True, histtype='step', label='Method b', color='red')

ax1.text(-0.35, 1.85, textstr1,bbox=dict(facecolor='white', edgecolor='black'), size=12)

ax1.legend(fontsize=10, fancybox=False, edgecolor='black')
ax1.set_ylabel('Frequency')
ax1.set_title('Trial 1')


# Second plot
ax2.hist(res_a2, bins=30, density=True, histtype='step', label='Method 1', color='blue')
ax2.hist(res_b2, bins=30, density=True, histtype='step', label='Method 2', color='red')

ax2.text(-0.5, 1.62, textstr2, bbox=dict(facecolor='white', edgecolor='black'), size=12)

ax2.legend(fontsize=10, fancybox=False, edgecolor='black')
ax2.set_title('Trial 2')

fig.text(0.5, -0.04, '$\Delta E$ [Joules]', ha='center', size=20)
plt.show()
```

### Διανυσματικά πεδία

Τέτοια είδους διαγράμματα είναι εξαιρετικά χρήσιμα όταν θέλουμε να οπτικοποιήσουμε το πως μεταβάλλεται ένα διανυσματικό μέγεθος (π.χ. η ταχύτητα ενός ρευστού σε διάφορα σημεία).

```python
w = 3
_ = np.linspace(-3,3,100)
X, Y = np.meshgrid(_,_)
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U**2 + V**2)
```

```python
fig, axes = plt.subplots(2,2,figsize=(5,5))
ax = axes[0][0]
ax.streamplot(X, Y, U, V)
ax = axes[0][1]
ax.streamplot(X, Y, U, V, color=speed)
ax = axes[1][0]
lw = 5*speed / speed.max()
ax.streamplot(X, Y, U, V, linewidth=lw)
ax = axes[1][1]
seed_points = np.array([[0,1], [1,0]])
ax.streamplot(X, Y, U, V, color=U, linewidth=2,
                      cmap='autumn', start_points=seed_points)
ax.grid()
```
