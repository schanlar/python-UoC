# Οπτικοποίηση δεδομένων με την βιβλιοθήκη ``Matplotlib``

**This notebook is heavily infuenced by the work of Luke Polson ([check here](https://github.com/lukepolson))**

Η ``Matplotlib`` είναι μια ευέλικτη και ευρέως χρησιμοποιούμενη βιβλιοθήκη σχεδίασης για την Python, σχεδιασμένη για να διευκολύνει την οπτικοποίηση και την ανάλυση δεδομένων. Η απλή διεπαφή που παρέχει, επιτρέπει στους χρήστες να δημιουργούν μια ευρεία γκάμα γραφημάτων και γραφημάτων υψηλής ποιότητας, από απλές γραφικές παραστάσεις έως πολύπλοκα γραφήματα και animations. 

Η Matplotlib λειτουργεί με βάση την έννοια των σχημάτων (`Figure`) και των αξόνων (`Axes`):

**Figure**:

* Ένα αντικείμενο τύπου ``Figure`` είναι το ανώτερο στην ιεραρχία και αποτελεί το γενικό "δοχείο" ή καμβά που θα περιέχει όλα τα στοιχεία ενός γραφήματος ή ενός συνόλου από επι μέρους γραφήματα (subplots), κείμενο, λεζάντες κτλ.


* Μπορούμε να δημιουργήσουμε ένα νέο αντικείμενο τύπου ``Figure`` χρησιμοποιώντας την εντολή ``plt.figure()``.


* Ένα αντικείμενο τύπου ``Figure`` παρέχει παραμέτρους όπως το μέγεθος, το χρώμα του υποβάθρου, τίτλο κτλ που μας επιτρέπουν να προσαρμόσουμε κάθε πτυχή στα δικά μας κριτήρια. Αυτοί οι παράμετροι μπορούν να τροποποιηθούν χρησιμοποιώντας εντολές όπως οι ``figsize``, ``facecolor``, και ``suptitle`` όπως θα δούμε στη συνέχεια.
    
    
**Axes**:

* Μέσα σε ένα αντικείμενο τύπου ``Figure``, μπορούμε να έχουμε ένα ή περισσοτέρα αντικείμενα τύπου ``Axes``. Ένα τέτοιο αντικείμενο είναι το δεύτερο στην ιεραρχία και αποτελεί την περιοχή στην οποία απεικονίζονται τα δεδομένα. 
    
    
* Μπορείτε να σκεφτείτε ένα αντικείνενο τύπου ``Axes`` ως ένα μεμονωμένο γράφημα μέσα σε ένα αντικείμενο τύπου ``Figure`` που όπως είπαμε αποτελεί τον καμβά.


* Μπορούμε να δημιουργήσουμε ένα νέο αντικείμενο τύπου ``Axes`` χρησιμοποιώντας εντολές όπως η ``plt.subplots()`` όπου δημιουργούμε πρώτα το αντικείμενο τύπου ``Figure`` και ένα ή περισσότερα αντικείμενα τύπου ``Axes`` με μία μόνο κλήση.

* Κάθε αντικείμενο ``Axes`` έχει δύο άλλα αντικείμενα τύπου ``Axis`` - ένα για τον άξονα-x και ένα για τον άξονα-y. Μέσω αυτών των στιγμιότυπων της κλάσης ``Axis`` μπορούμε να διαχειριστούμε τα ονόματα σε κάθε άξονα, τις υποδιαιρέσεις κτλ.



Είτε πρόκειται για επιστημονική έρευνα, για εξερεύνηση δεδομένων ή για δημιουργία γραφημάτων που προορίζονται για κάποια δημοσίευση ή αναφορά, η βιβλιοθήκη ``Matplotlib`` παρέχει ένα ισχυρό σύνολο εργαλείων για τη δημιουργία σαφών και κατατοπιστικών απεικονίσεων στη γλώσσα προγραμματισμού Python. Για περισσότερες πληροφορίες σχετικά με την εγκατάσταση της βιβλιοθήκης ή παραδείγματα χρήσης της μπορείτε να επισκεφθείτε την επίσημη [ιστοσελίδα](https://matplotlib.org/).

## Quick plotting

Η βιβλιοθήκη ``Matplotlib`` μας παρέχει διάφορες συναρτήσεις για την δημιουργία διαφορετικών τύπων γραφημάτων. Για παράδειγμα, την συνάρτηση ``plot()`` για την δημιουργία γραφημάτων κάποιας καμπύλης, την συνάρτηση ``scatter()`` για την δημιουργία γραφημάτων διασποράς, την συνάρτηση ``bar()`` για την δημιουργία ραβδογραμμάτων κτλ. 

Όλες αυτές οι συναρτήσεις δρουν πάνω σε ένα αντικείμενο τύπου ``Axes``. Μπορούμε εάν το επιθυμούμε να δηλώσουμε ρητά τα αντικειμένα τύπου ``Figure`` και ``Axes`` πάνω στα οποία θα κατασκευάσουμε τα γραφήματά μας ή μπορούμε να αφήσουμε την ``Matplotlib`` να δημιουργήσει αυτόματα αυτά τα αντικείμενα για εμάς. Υπάρχουν προτερήματα και μειονεκτήματα και στις δύο μεθόδους. 

Γενικά, όταν θέλουμε να κάνουμε ένα γρήγορο γράφημα για να δούμε απλά την μορφή των δεδομένων μας και να κάνουμε μία βασική ανάλυση, συνήθως αφήνουμε την ``Matplotlib`` να διαχειριστεί το κομμάτι της δημιουργίας των αντικειμένων ``Figure`` και ``Axes``. Όταν μας ενδιαφέρει να δημιουργήσουμε ένα πιο πολύπλοκο και ποιοτικό γράφημα τότε, πολλές φορές, η ρητή δημιουργία αυτών των αντικειμένων μας δίνει περισσότερες ελευθερίες ως προς την διαχείριση των γραφημάτων (δες και Παράρτημα Α).

````python
# Implicitly create the Figure and Axes objects - Matplotlib  automatically handles this

# Create a Figure object of 8in x 4in dimension
plt.figure(figsize=(8,4))

# Get the current Axes object - if it doesn't exist Matplotlib creates it
plt.gca()

# Use the newly created Axes object to modify some parameters
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("This is a title")

plt.show()
````

````python
# Explicitly create the Figure and Axes objects

# Figure object
fig = plt.figure(figsize=(8,4))

# Axis object
ax = plt.gca()

# Use the Axes object to modify some parameters - notice the slighly different syntax
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.set_title("This is a title")

plt.show()
````

Συνήθως χρησιμοποιούμε τόσο απλά γραφήματα απλά για να εξετάσουμε τα δεδομένα μας. - **Πολλαπλές καμπύλες σε ένα γράφημα**

````python
# Create the data for fitting curves
y1_fit_values = np.sin(x_values)
y2_fit_values = np.cos(x_values)
````

````python
plt.figure(figsize=(8,3))
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Voltage [V]', fontsize=16)
plt.title('Voltage vs Time', fontsize=16)

# Plot only the data points
_ = plt.plot(x_values, y1_values, marker='o', ms=5, ls=' ', c='red', label='sin data')
_ = plt.plot(x_values, y2_values, marker='o', ms=5, ls=' ', c='blue', label='cos data')

# Plot the fit curves
_ = plt.plot(x_values, y1_fit_values, ls='--', lw=2, c='red', label='sin (fit)')
_ = plt.plot(x_values, y2_fit_values, ls='-.', lw=2, c='blue', label='cos (fit)')

_ = plt.legend(loc='lower left', fontsize=15)
````

### Γραμμοσκίαση περιοχής

Η γραμμοσκίαση περιοχών σε ένα γράφημα μπορεί να αποδειχτεί μία πολύ χρήσιμη τεχνική καθώς επιτρέπει την εύκολη οπτικοποίηση των περιοχών για τις οποίες ισχύει κάποια συνθήκη.

````python
# The size of the figure
plt.figure(figsize=(8,3))

# Add axes labels 
plt.xlabel('Time [s]', fontsize=15)
plt.ylabel('Voltage [V]', fontsize=15)
plt.title('Voltage vs Time', fontsize=16)

# The axes boundary
plt.xlim()
plt.ylim(-1.2, 2.5)

# Plot only the data points
_ = plt.plot(x_values, y1_values, marker='o', ms=2, ls=' ', c='red', label='sin data')
_ = plt.plot(x_values, y2_values, marker='o', ms=2, ls=' ', c='blue', label='cos data')

# Plot the fit curves
_ = plt.plot(x_values, y1_fit_values, ls='--', lw=1, c='red', label='sin (fit)')
_ = plt.plot(x_values, y2_fit_values, ls='-.', lw=1, c='blue', label='cos (fit)')

# Fill between the lines
# _ = plt.fill_between(x_values, y1_fit_values, y2_fit_values, color='purple', alpha=0.25)

_ = plt.fill_between(x_values, y1_fit_values, y2_fit_values, color='orange', alpha=0.25, 
                     where=(y1_fit_values > y2_fit_values))

_ = plt.fill_between(x_values, y1_fit_values, y2_fit_values, color='black', alpha=0.25, 
                     where=(y1_fit_values < y2_fit_values))


# Add the legend
_ = plt.legend(loc='upper center', fontsize=13, ncol=2)
````

### Ιστογράμματα (histograms)

Τα ιστογράμματα είναι χρήσιμα όταν θέλουμε να οπτικοποιήσουμε τον τρόπο με τον οποίο κατανέμονται τα δεδομένα μας, όταν τα δεδομένα μας βρίσκονται εντός κάποιων ορίων.

Το ιστογράμματα θυμίζουν αρκετά τα ραβδογράμματα, αλλά τώρα τα δεδομένα ομαδοποιούνται σε "κάδους" (bins) αντί κάθε ράβδος να αναπαριστά μία συγκεκριμένη τιμή.

````python
# Create some data
data = np.random.randn(1000)*0.2 + 0.4
````

````python
plt.figure(figsize=(8,3))
_ = plt.hist(data)
````

- **Πολλαπλά ιστογράμματα σε ένα γράφημα**

Η προσθήκη πολλαπλών ιστογραμμάτων σε ένα γράφημα μπορεί πολλές φορές να μην είναι ιδανική καθώς το ένα ιστόγραμμα μπορεί να επικαλύπτει μεγάλη περιοχή του άλλου.

````python
data2 = np.random.randn(1000)*0.2 + 0.4
````

````python
plt.figure(figsize=(8,3))
_ = plt.hist(data, bins=30, density=True)
_ = plt.hist(data2, bins=30, density=True)
````

````python
plt.figure(figsize=(8,3))
_ = plt.hist(data, bins=30, density=True, histtype='step')
_ = plt.hist(data2, bins=30, density=True, histtype='step')
````

Το ιστόγραμμα επιστρέφει 2 πίνακες που συχνά είναι χρήσιμοι:

1. Τον αριθμό των τιμών που περιέχει κάθε κάδος.

## Παράρτημα Α (Implicit vs Explicit Figure and Axes Creation)

### Key Differences:
1. Implicit vs Explicit Figure and Axes Creation:

 - ``plt.plot`` implicitly manages figure and axis creation.

 - ``ax.plot`` requires explicit creation of a figure and axis before plotting.


2. Current Axis vs. Specified Axis:

 - ``plt.plot`` uses the current axis (``plt.gca()``) by default but creates one if it doesn't exist.
 
 - ``ax.plot`` explicitly specifies the axis on which to plot, which is useful when working with multiple subplots.
 
 
3. Object-Oriented vs. State-Based:

 - ``plt.plot`` is state-based and relies on the current state to determine where to plot.

 - ``ax.plot`` is object-oriented, allowing for more fine-grained control over individual axes.


### When to Use Which:

1. Simple Plotting:
For simple, single plots, ``plt.plot`` is often more convenient.


2. Advanced Plotting:
For more complex visualizations, especially when dealing with multiple subplots, it's recommended to use the ``ax.plot`` approach for explicit control.


3. Consistency:
If you are working on a script or project with multiple developers, using ``ax.plot`` consistently can lead to clearer and more maintainable code.

### Διανυσματικά πεδία

Τέτοια είδους διαγράμματα είναι εξαιρετικά χρήσιμα όταν θέλουμε να οπτικοποιήσουμε το πως μεταβάλλεται ένα διανυσματικό μέγεθος (π.χ. η ταχύτητα ενός ρευστού σε διάφορα σημεία).

````python
w = 3
_ = np.linspace(-3,3,100)
X, Y = np.meshgrid(_,_)
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U**2 + V**2)
````

````python
fig, axes = plt.subplots(2,2,figsize=(5,5))
ax = axes[0][0]
ax.streamplot(X, Y, U, V)
ax = axes[0][1]
ax.streamplot(X, Y, U, V, color=speed)
ax = axes[1][0]
lw = 5*speed / speed.max()
ax.streamplot(X, Y, U, V, linewidth=lw)
ax = axes[1][1]
seed_points = np.array([[0,1], [1,0]])
ax.streamplot(X, Y, U, V, color=U, linewidth=2,
                      cmap='autumn', start_points=seed_points)
ax.grid()
````
